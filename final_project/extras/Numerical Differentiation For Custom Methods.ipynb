<p>This document is used to simulate differentiation techniques on custom math functions which users themselves will provide in the beginning.

```python
# Imports.
import math
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3>First, Specify The Custom Math Function In The Cell Below :</h3>

```python
# Wrap the math function in a lambda statement :
func = lambda x: np.sin(x)
```

```python
# For Functions Involving Trigonometric Operations, Use The Sample Array Below To Construct The X - Axis For The Line Plot :
trig_x_axes = np.arange(0, math.pi*2, 0.05)
```

```python
# Non-Trigonometric Methods Will Implement A Sample List Of Integers Below For Its X - Axis :
nontrig_x_axes = [i for i in range(1, 100 + 1)]
```

<h3>Let's Try To Compare That Function With Its Own Derivative :</h3>

```python
plt.figure(figsize = (12, 5))
plt.plot(trig_x_axes, func(trig_x_axes), label = "F(X) = Sin(X)")
plt.plot(trig_x_axes, derivative(func, trig_x_axes, dx = 0.5), label = "F(X) = Cos(X)")
plt.legend()
plt.grid(True)
plt.show()
```
